```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
from IPython import display
import os
import uuid
import time
import cv2

# load custom model
model = torch.hub.load('ultralytics/yolov5','custom', path='yolov5/runs/train/exp8/weights/last.pt')

# video testing
camera = cv2.VideoCapture("1.mp4")
while camera.isOpened():
    ret, frame = camera.read()
    
#     make detections
    predictions = model(frame)
    print(predictions.xyxy)
    
    cv2.imshow('Red Card Detection', np.squeeze(predictions.render()))
    
    if cv2.waitKey(10) & 0xFF==ord('q'):
        break
        
camera.release()
cv2.destroyAllWindows()
```
